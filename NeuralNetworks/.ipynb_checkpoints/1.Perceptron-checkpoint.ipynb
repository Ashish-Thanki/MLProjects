#### Peceptron is the simpliest artificial neural network architectures, it uses the threshold logic unit (TLU). 

It can be used to classify multiple linear classification problems. It uses the weighted sum of its inputs and applies a step function. The step function in this case is the Heaviside Step Function, it is very similiar to the Logstic Regression function, where negative results will yield classification to the negative class and postive results that exceed the threshold value will be classified in the positive class. 

Computing the outputs of a fully connected layer.
$h_{W, b}(X) = \phi(XW+b)$
where:

$X$, where each row represents an instance and a column represents a feature

$W$, weight matrix, without the bias term, one row per <b>input</b> neuron and one column per layer <b>neuron in that layer</b>.

$\phi$ is the <b>activation function</b>, it the TLU case, this is a step function but there are many other types.

## What are the types of activation functions?

- Step 
- Sigmoid $1/ (1+e^{-\theta x})$
- Rectified Linear Unit max($0, z$)
- Hyperbolic Tangent Function - Tanh($z$) = 2$\sigma(2z) - 1$

```python
import numpy as np
#sigmoid function
z = np.linspace(-3, 3, 100)
y_sig = 1/(1+ np.exp(-z))

plt.title('Sigmoid Function')
plt.grid(True)
plt.plot(z, y_sig, label= 'Sigmoid')
```

```python
#Hyperbolic Tangent Function 

y_tanh = np.tanh(z)


plt.title('Hyperbolic Tangent Function')
plt.grid(True)
plt.plot(z, y_tanh, label= 'Tanh')
```

```python
# Rectified Linear Unit Function

y_relu= np.maximum(0, z) #returns the largest value out of 0 and z

plt.title('Rectified Linear Unit Function')
plt.grid(True)
plt.plot(z, y_relu, label='ReLU')
```

```python
#Step Function, using https://numpy.org/doc/stable/reference/generated/numpy.sign.html 

y_step = np.sign(z)

plt.title('Step')
plt.grid(True)
plt.plot(z, y_step, label='Step')
```

```python
#Combine into one graph

plt.figure(figsize=(14, 6,))
plt.axis([-3,3, -1.1, 1.1])
plt.grid()
plt.title('Activation Functions', fontsize= 16, fontweight= 'bold')

plt.plot(z, y_sig,'g--', label= 'Sigmoid', )
plt.plot(z, y_tanh,'b--', label= 'Tanh', )
plt.plot(z, y_relu,'r', label='ReLU', )
plt.plot(z, y_step,'m', label='Step', )

plt.legend(fontsize=14,loc= 'center right')
```

Conversly, if a linear activation function were to be incorporated then the associated deep stack layers will produce linear functions also (or a continous variable, possibly zero). This makes it impossible for a deep neural network to 'converge' to complex problems and the deep neural network is equivalent a network with just a single layer.


# Multilayer Perceptron
You can use MLP for Regression and Classification problems - <i>you can guess multiple values or classes!!</i>

What is the typical architecture of neural network for <b>regression</b>?
1. Input Neurons (One per feature)
2. Hidden layers (i.e. all layers between the input layer and output layers)
3. Neurons per hidden layer which uses their own hidden activation function (can be anywhere between 10 to 100 neurons per layer)
4. Output Neurons - 1 per prediction dimension, which uses it's own output activation function. Using ReLU and Step for positive outputs and, Sigmoid and Tanh for predictions expected to fall within a given range.
5. Note, the loss function is the mean square error, however, if there are loads of outliers then you should use mean absolute error. Interestingly, you can apply both by using Huber Loss. The function is given a threshold and applies the mean squared error to values that fall below it and mean absolute error for values above. The outcome being a faster convergence. 

What is the typical architecture of neural network for <b>classification</b>?

1. Input Neurons - Same as regression
2. Hidden layers - Same as regression
3. Neurons per hidden layer - Same as regression
4. Output Neurons - 1 output Neuron, can be Softmax Activation Function. This is because all the neurons use logistic regression prior to the output neuron and the Softmax function ensures all the outputs are between 0 and 1 - and add up to 1 which is not always the case because each neuron will use its own activation function. 
5. Note, the loss function is the cross-entropy loss.

#### Wide and Deep is a type of Nonsequential Neural Networks
This is where a network is created, where it uses the wide path for simple patterns (i.e. a path from the input to the Concatenate layer) and the deep neural network path for deep patterns. 

##### Building Wide and Deep Model

```python
tf.keras.backend.clear_session()#keep the memory clean of the previous model
```

```python
input_ = keras.layers.Input(shape= X_train.shape[1:]) #specify the input shape of one instance
hidden1 = keras.layers.Dense(30, activation= 'relu')(input_) #tell keras 30 neurons are to be connected to the input layer
hidden2 = keras.layers.Dense(30, activation= 'relu')(hidden1) #create 2nd hidden layer & pass output from 1st hidden layer
concat = keras.layers.Concatenate()([input_, hidden2]) #output from hidden layer 2 concated with input
output = keras.layers.Dense(1)(concat) #create the output layer with a single neuron 
model_wide = keras.models.Model(inputs= [input_], outputs= [output]) #specify which inputs and outputs to use
```

## Dividing subsets of data so that some features go through the wide path and some features go through the deep path

```python
#create input layers
input_A = keras.layers.Input(shape=[5], name='Wide_Input') 
input_B = keras.layers.Input(shape=[6], name='Deep_Input')

#create hidden layers
hidden1 = keras.layers.Dense(30, activation= 'relu')(input_B)
hidden2 = keras.layers.Dense(30, activation= 'relu')(hidden1)

#create wide path
concat = keras.layers.Concatenate()([input_A, hidden2])

#create output layers
output = keras.layers.Dense(1, name='output')(concat)

model_wide_partial = keras.models.Model(inputs=[input_A, input_B], outputs= [output])
```

```python
#create neural network

input_A = keras.layers.Input(shape=[5], name= 'wide_input')
input_B = keras.layers.Input(shape=[6], name= 'deep_input')

hidden1 = keras.layers.Dense(30, activation= 'relu')(input_B)
hidden2 = keras.layers.Dense(30, activation= 'relu')(hidden1)

concat = keras.layers.Concatenate()([input_A, hidden2])
output = keras.layers.Dense(1, name= 'name_ouput')(concat)
aux_output = keras.layers.Dense(1, name='aux_output')(hidden2)

model = keras.models.Model(inputs=[input_A, input_B], outputs= [output, aux_output])
```

### The Sequential and Functional API have been <u>Declarative</u>.
You declare which layers you want and how they are connected and are not concerned with how the function actually does it (see declarative vs imperative below). This is easy for us to debug, among many other benefits including: displaying into a static graph, analyzing, saving, reusing and sharing. 

But what if we want models that contain loops, varying shapes, conditional branches and more <b>dynamic</b> flexibility?
### Subclassing API is an <u>Imperative</u> programming style. 
You create a class which uses keras.Model as a subclass, create layers in the constructor (i.e. initialize function) and perform all the computations in the call() method. 

The class below does the exact same thing as the Wide and Deep model we created earlier, we then can compile, evaluate and make predictions using the imperative class created.

```python
class WideAndDeepModel(keras.models.Model):
    def __init__(self, units=30, activation= 'relu', **kwargs):
        super().__init__(**kwargs) #use base class to assign kwargs
        #https://www.programiz.com/python-programming/methods/built-in/super
        #https://www.programiz.com/python-programming/args-and-kwargs
        
        self.hidden1 = keras.layers.Dense(units, activation=activation)
        self.hidden2 = keras.layers.Dense(units, activation=activation)
        self.main_output = keras.layers.Dense(1)
        self.aux_output = keras.layers.Dense(1)
        
    def call(self, inputs):
        input_A, input_B = inputs
        hidden1 = self.hidden1(input_B)
        hidden2 = self.hidden2(hidden1)
        concat = keras.layers.concatenate([input_A, hidden2])
        main_output = self.main_output(concat)
        aux_output = self.aux_output(hidden2)
        return main_output, aux_output


model = WideAndDeepModel(30, activation= 'relu')
```

# Long training hours
Sometimes, models may take several hours to learn. We should definitely save the model at the end of training but also __during__ training too! We can save the model at several checkpoints at regular intervals to avoid being left nothing if our computer crashes. Fortunetly, the .fit() method has a callbacks argument to allow you save after every epoch, for example. 

```python
tf.keras.backend.clear_session()
```

```python
model = keras.Sequential([
    keras.layers.Dense(30, activation='relu', input_shape=[8]),
    keras.layers.Dense(30, activation= 'relu'),
    keras.layers.Dense(1)
])
```

##### by setting save_best_only= True, we save the model that performs best in the validation data.

### Saving the best model

```python
model.compile(loss='mse', optimizer=keras.optimizers.SGD(lr=0.001))

#checkpoint as specified below
checkpoint_cb = keras.callbacks.ModelCheckpoint('my_keras_model.h5', save_best_only=True)

#specify the callbacks with the checkpoint above
history = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10,
                    callbacks=[checkpoint_cb])

#loads best model
model = keras.models.load_model('my_keras_model.h5')
print(model.evaluate(X_test, y_test))
```

#### Early stopping can be incorporated by using the EarlyStopping callback
It stops training when the model does not improve on the validation set. The Patience argument is the number of epochs required, where there is no progress on the validation set, to interrupt training, where the best model is returned.

### You can combine both Saving Checkpoints and Early Stopping to prevent your computer crashing and avoid wasting time & resources.

```python
early_cb = keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)#set patience and return best weights= True

history = model.fit(X_train, y_train, epochs=100,
                   validation_data=(X_val, y_val),
                   callbacks=[checkpoint_cb, early_cb])
```

```python
print(model.evaluate(X_test, y_test))
```

# Success! It worked and saved the best model as well as using early stopping! 
Callbacks can be used to help restore your best model, just before you computer crashes. 
There are other types of callbacks and even remote loggers! See Keras Documentation (it's an easy read) - https://keras.io/api/callbacks/ 

You can create custom callbacks by creating a class with keras.callbacks.Callback as your base class.

```python
class PrintValidTrainRatioCallback(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):#specifically have to state this function otherwise this will not work
        print('\nval/train: {:.2f}'.format(logs['val_loss'] / logs['loss']))
```

```python
val_train_ratio_cb = PrintValidTrainRatioCallback()
history = model.fit(X_train, y_train, epochs=3,
                   validation_data=(X_val, y_val), 
                   callbacks=[val_train_ratio_cb])
```

## Callback Extra Control:
As with the PrintValidTrainRatioCallback class, you need to define a function inside called on_epoch_end specifically, otherwise, the function will not print what you require.

Other functions you can define are:
- on_train_begin() & on_train_end()
- on_epoch_begin() & on_epoch_end()
- on_batch_begin() & on_epoch_end()

These could be useful for debugging and evaluation, by calling the evaluate method you can create custom functions such as:
- on_test_begin() & on_test_end()
- on_test_batch_begin() & on_test_batch_end()

and finally, you can create callbacks for predictions, through calling .predict() method, and create custom functions such as:
- on_predict_begin() & on_predict_end()
- on_predict_batch_begin() & on_predict_batch_end()

## Fine Tuning Neural Network Hyperparameters
Evidentally, there are loads of hyperparameters to optimize, from number of neurons, layers, activation functions, neural network architecture, weight initialization and many many more.

To help find the best hyperparameters, you can use <b>GridSearchCV and RandomizedCV</b> but you need to create a function that will build the neural network and compile it with hyperparameters as it's input. 

```python
keras.backend.clear_session()
```

```python
def build_model(n_hidden=1, n_neurons= 30, learning_rate=3e-3, input_shape=[8]):
    #sequential model
    model = keras.models.Sequential()
    
    #add an input lyer
    model.add(keras.layers.InputLayer(input_shape= input_shape))
    
    #add number of hidden layer
    for layer in range(n_hidden):
        model.add(keras.layers.Dense(n_neurons, activation='relu'))
    
    #output layer
    model.add(keras.layers.Dense(1))
    
    #define optimizer and it's learning rate
    optimizer= keras.optimizers.SGD(lr= learning_rate)
    
    #compile model
    model.compile(loss= 'mse', optimizer= optimizer)
    
    return model

# you should also set the default values, just like Sci-Kit Learn
```

```python
# Now we need to create a KerasRegressor using the scikit_learn wrapper and load our model into it

keras_reg = keras.wrappers.scikit_learn.KerasRegressor(build_model)
```

```python
#fit the model 

keras_reg.fit(X_train, y_train, epochs= 100,
             validation_data= (X_val, y_val),
             callbacks= [keras.callbacks.EarlyStopping(patience=10)])
```

```python
mse_test = keras_reg.score(X_test, y_test)
mse_test #opposite of the MSE, higher value is better
#Sci-kit learn prefers scores not losses, unlike neural networks
```

__Warning__ the next cell will take a long time to run, I have reduced the hyperparameters because of this. 

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import reciprocal

params_distribs = {
    'n_hidden' : [1, 3],
    'n_neurons': np.arange(50, 100, 10),
    'learning_rate': reciprocal(0.001, 0.003)
}

rnd_search_cv = RandomizedSearchCV(keras_reg, params_distribs, n_iter=5, cv=3, verbose=1, n_jobs=-1)

rnd_search_cv.fit(X_train, y_train, epochs= 100,
                 validation_data= (X_val, y_val),
                 callbacks= [keras.callbacks.EarlyStopping(patience=5)])
```

```python
#create a function that can go into Sci-kit Learn's Randomized Search CV

def model_build(first_hidden_neurons= 300, hidden_neurons = 100, n_hidden=2,
                learning_rate= 0.001, input_shape=[28,28]):
    #build model
    model = keras.models.Sequential()
    
    #add layers
    model.add(keras.layers.Flatten(input_shape= input_shape))
    
    #first layer
    model.add(keras.layers.Dense(first_hidden_neurons, activation= 'relu'))
    
    for i in range(n_hidden):
        model.add(keras.layers.Dense(hidden_neurons, activation='relu'))
    
    #output layer
    model.add(keras.layers.Dense(10, activation= 'softmax'))
    
    #compile model
    optimizer= keras.optimizers.SGD(lr=learning_rate)
                                    
    model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer, metrics= ['accuracy'])
    return model
    
```

```python
# Now we need to create a KerasRegressor using the scikit_learn wrapper and load our model into it
keras_reg = keras.wrappers.scikit_learn.KerasClassifier(model_build)
```

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
from scipy.stats import reciprocal

params_distr = {
    'learning_rate': reciprocal(0.001, 0.003)
}

random_mnist_clf = RandomizedSearchCV(keras_reg, param_distributions=params_distr,
                                      n_iter=5, cv=3, n_jobs=-1)

random_mnist_clf.fit(X_train, y_train, epochs= 100,
                    validation_data= (X_valid, y_valid), 
                    callbacks= [keras.callbacks.EarlyStopping(patience=10)])
```

```python
class ExponentialLearningRate(keras.callbacks.Callback):
    def __init__(self, factor):
        self.factor = factor
        self.rates = []
        self.losses = []
    
    def on_batch_end(self, batch, logs):
        self.rates.append(keras.backend.get_value(self.model.optimizer.lr))
        self.losses.append(logs['loss'])
        keras.backend.set_value(self.model.optimizer.lr, self.model.optimizer.lr * self.factor)
```

```python
keras.backend.clear_session()
log_model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.Dense(300, activation= 'relu'),
    keras.layers.Dense(100, activation= 'relu'),
    keras.layers.Dense(10, activation= 'softmax')
])
log_model.compile(loss= keras.losses.sparse_categorical_crossentropy, metrics= ['accuracy'],
                 optimizer= keras.optimizers.SGD(lr= 1e-3))

expon_lr = ExponentialLearningRate(factor=1.005)
```

```python
log_history = log_model.fit(X_train, y_train, epochs=1, 
             validation_data=(X_valid, y_valid), 
             callbacks=[expon_lr])
```

```python
#plot the log curve for the log loss and 

plt.figure(figsize=(12, 6))
plt.gca().set_xscale('log')
plt.plot(expon_lr.rates, expon_lr.losses,)
plt.hlines(min(expon_lr.losses), min(expon_lr.rates), max(expon_lr.rates))
plt.axis([min(expon_lr.rates), max(expon_lr.rates), 0.5, 2.5])

plt.axvline(x=0.0024, c='r')
plt.xlabel('Learning Rate')
plt.ylabel('Loss')
plt.title('Learning Curve - for different Learning Rates')
```
